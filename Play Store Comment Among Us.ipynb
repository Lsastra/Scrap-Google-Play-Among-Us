```python
import time
import re
import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By

from tqdm import tqdm
```

# 1. Scrap Apple Play Games (Winata Syahputra)

```python
WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.XPATH, "//span[contains(@class,'RveJvd snByac')]")))
while True:
    try:
        WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "//span[contains(@class,'RveJvd snByac')]"))).click()
    except TimeoutException:
        break
```

```python
page_url = "https://play.google.com/store/apps/details?id=com.innersloth.spacemafia&showAllReviews=true"
driver = webdriver.Chrome()
driver.get(page_url)

SCROLL_PAUSE_TIME = 2
ROUND = 5
CYCLES = 5

html = driver.find_element_by_tag_name('html')
html.send_keys(Keys.PAGE_DOWN)  
html.send_keys(Keys.PAGE_DOWN)  
time.sleep(SCROLL_PAUSE_TIME * 3)

for m in tqdm (range(ROUND)):
    for i in tqdm (range(CYCLES)):
        html.send_keys(Keys.END)
        time.sleep(SCROLL_PAUSE_TIME)
    WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.XPATH, "//span[contains(@class,'RveJvd snByac')]"))).click()
```

```python
v_komen = driver.find_elements_by_xpath('//*[@class = "UD7Dzf"]/span[1]')
v_pengguna = driver.find_elements_by_xpath('//*[@class="bAhLNe kx8XBd"]/span[@class = "X43Kjb"]')
v_tanggal = driver.find_elements_by_xpath('//*[@class="bAhLNe kx8XBd"]/div/span[@class="p2TkOb"]')
v_bintang = driver.find_elements_by_xpath('//*[@class="nt2C1d"]/div[@class = "pf5lIe"]/div')
```

```python
pengguna = []
tanggal = []
komen = []

number_of_items = len(v_pengguna)
for berita in tqdm (range(number_of_items)):
    pengguna.append(str(v_pengguna[berita].text))
    tanggal.append(str(v_tanggal[berita].text))
    komen.append(str(v_komen[berita].text))
```

```python
bintang = []
for alamat in tqdm(v_bintang):
    bintang.append(alamat.get_attribute('aria-label'))
```

```python
df = pd.DataFrame()
df["Bintang"] = bintang
df["Pengguna"] = pengguna
df["Komen"] = komen
df["Tanggal"] = tanggal
```
